```python
from selenium import webdriver
import pandas as pd
import time
```

```python
driver = webdriver.Chrome(executable_path = '/Users/aayushsharma/Downloads/chromedriver')
```

```python
driver.get('https://www.imdb.com/list/ls002913270/')
```

```python
celebs = driver.find_elements_by_xpath("//*[@class = 'lister-item mode-detail']")
```

```python
Celebrity_data = []

column_names = ['Name', 'Job Catogories', 'Born', 'Star Sign', 'Height', 'Nicknames', 'Bio', 'Image Link']

for i in celebs[:1]:
    
    base_url = 'https://www.imdb.com'
    data = BeautifulSoup(driver.page_source, 'lxml')
    celeb = data.find_all(class_ = 'lister-item-header')
    for i in celeb:
        name = i.text[5:].strip()
        
        if name == 'Rani Mukherjee' or name == 'Jaaved Jaffery' or name == 'Parineeti Chopra' or name == 'Omi Vaidya':
            continue
            
        else:
            
            driver.get(base_url + str(i.a['href']))
            time.sleep(2)
            
            Job_catogories = driver.find_element_by_id('name-job-categories').text
            
            Star_sign = driver.find_element_by_id('dyk-star-sign').text[10:].strip()
            
            data = BeautifulSoup(driver.page_source, 'lxml')

            bio = data.find(class_ = 'see-more inline nobr-only')
            driver.get(base_url + str(bio.a['href']))
            time.sleep(2)

            data = BeautifulSoup(driver.page_source, 'lxml')

            Name = data.find(class_ = 'parent').text.strip()

            Born = data.find(class_ = 'even').text[4:].strip()

            Nicknames = data.find(class_ = 'odd').text[9:].strip()

            Height = data.find_all(class_ = 'even')[1].text[7:].strip()

            Bio = data.find(class_ = 'soda odd').text.strip()

            Image_link = data.find(class_ = 'poster')['src']

            Celebrity_data.append([Name, Job_catogories, Born, Star_sign, Height, Nicknames, Bio, Image_link])
        
df = pd.DataFrame(Celebrity_data, columns = column_names)
            
```
